# Контрольні питання

## 1) Жадібні алгоритми - це алгоритми, які вирішують проблему шляхом вибору на кожному кроці найоптимальнішого варіанту. Вони роблять локально найкращий вибір у кожній точці з метою здійснення глобально оптимального рішення.

## 2) Префіксний код - це система кодування, в якій жоден код не є префіксом іншого коду. Код, що використовується у коді Гафмена, є префіксним, що дозволяє однозначно розкодувати закодовані дані.

## 3) Купа - це деревоподібна структура даних, де кожен вузол має значення, більше або менше значень його дітей. Черга з пріоритетами може бути побудована на основі купи, де кожен елемент має пріоритет та розташовується так, щоб елементи з вищим пріоритетом знаходилися вгорі.

## 4) Стиснення даних - це процес скорочення обсягу даних з метою економії простору або ресурсів мережі. Воно використовується для зменшення обсягу даних для зберігання, передачі або обробки. Його основні переваги - ефективне використання простору, зменшення часу передачі та економія ресурсів.

## 5) Для стиснення даних за допомогою алгоритму кодування Гафмена потрібно - побудувати дерево кодування Гафмена на основі частот входження символів, створити таблицю кодів, де кожному символу буде присвоєно його код, закодувати вихідне повідомлення, замінивши кожен символ на його відповідний код.

## 6) Основні обмеження та недоліки алгоритму кодування Гафмена включають - великий обсяг додаткових даних для зберігання таблиці кодів, потребу у додатковому часі на створення таблиці кодів та закодування даних, відносно низьку продуктивність на коротких повідомленнях, можливі покращення можуть включати оптимізацію побудови таблиці кодів та використання адаптивних методів створення дерев кодування.

## 7) Альтернативні методи стиснення даних, які можуть конкурувати з алгоритмом Гафмена, включають алгоритми Лемпеля-Зіва, а також методи стиснення на основі архівування, такі як ZIP або RAR.

## 8) Практичні застосування алгоритмів стиснення даних, зокрема алгоритму Гафмена, включають зменшення обсягу даних для зберігання на пристроях з обмеженим простором, економію пропускної здатності мережі під час передачі даних через мережу, а також збільшення швидкості передачі даних і підвищення продуктивності систем зберігання та обробки даних.
